```python
1. What is unsupervised learning in the context of machine learning?

Ans- Unsupervised learning is a machine-learning method where algorithms discover hidden patterns or groupings in unlabeled data—without guidance or predefined labels. Common techniques include clustering, association rule mining, and dimensionality reduction. It excels at exploring data insights but lacks clear evaluation metrics.
```

```python
2. How does K-Means clustering algorithm work?

Ans- K‑Means clustering groups data into K clusters using a simple 4-step loop:
* Pick K random centroids from the data.
* Assign each point to its nearest centroid (using Euclidean distance).
* Update centroids to the average of all assigned points.
* Repeat steps 2–3 until assignments stop changing or max iterations reach.
```

```python
3. Explain the concept of a dendrogram in hierarchical clustering?

Ans- A dendrogram is a tree-like diagram that shows how clusters merge in hierarchical clustering. Each leaf is an individual data point. The height of a branch indicates how far apart clusters are before merging.
```

```python
4. What is the main difference between K-Means and Hierarchical Clustering?

Ans- The main difference between K-Means and Hierarchical Clustering:
* K‑Means:
 * Need to choose K (number of clusters) beforehand.
 * Fast, works well on large datasets.
 * Best for round, equal-sized clusters.
 * Sensitive to outliers and initialization.

* Hierarchical:
 * No need to pick cluster count in advance.
 * Builds a tree (dendrogram) you can cut at any level.
 * Can handle varied shapes, better for small data.
 * Slower (needs lots of memory/time). Irreversible merges.
```

```python
5. What are the advantages of DBSCAN over K-Means?

Ans- The advantages of DBSCAN over K-Means:
* No need to specify the number of clusters: DBSCAN automatically determines the number of clusters based on data density, unlike K-Means, which requires the number of clusters (K) to be predefined.
* Handles noise and outliers: DBSCAN can identify and exclude noise points, making it more robust to outliers and anomalies in data.
* Detects clusters of arbitrary shapes: DBSCAN is not limited to spherical clusters and can detect clusters of any shape, including elongated or irregularly shaped clusters.
* Scalability: DBSCAN is highly scalable and can handle large datasets efficiently, especially with spatial indexing structures like KD-trees or R-trees.
```

```python
6. When would you use Silhouette Score in clustering?

Ans- The Silhouette Score is a metric used to evaluate the quality of clusters in unsupervised learning. It measures how similar each point is to its own cluster (cohesion) compared to other clusters (separation). The score ranges from -1 to +1:
* +1: Points are well-clustered.
* 0: Points are on or near the boundary between clusters.
* Negative values: Points may be incorrectly assigned to clusters.
```

```python
7. What are the limitations of Hierarchical Clustering?

Ans- The limitations of Hierarchical Clustering are:
* High Computational Complexity:

The algorithm has a time complexity of O(n³), making it computationally expensive and less suitable for large datasets.
* Sensitivity to Noise and Outliers:

Hierarchical clustering can be sensitive to noise and outliers, which can distort the clustering results.
* Irreversible Merging:

Once clusters are merged, the decision cannot be undone, which can lead to suboptimal clustering if early decisions were incorrect
* Difficulty in Handling Non-Convex Shapes:

The algorithm may struggle to identify clusters with non-convex shapes or varying densities.
* Challenges with High-Dimensional Data:

In high-dimensional spaces, hierarchical clustering can face challenges due to the curse of dimensionality, where data points become sparse, and distance measures become less meaningful.
```

```python
8. Why is feature scaling important in clustering algorithms like K-Means?

Ans- Feature scaling is important in K-Means because it makes sure all features are on the same scale. This prevents features with bigger numbers from dominating the distance calculations, so clusters form correctly.
```

```python
9. How does DBSCAN identify noise points?

Ans- DBSCAN identifies noise points as data points that do not belong to any cluster. These points have fewer neighbors than the minimum required (MinPts) within a given radius (epsilon), so they are considered outliers or noise.
```

```python
10. Define inertia in the context of K-Means?

Ans- Inertia in K-Means is the sum of squared distances between each data point and its nearest cluster center.  
     It measures how tight the clusters are—lower inertia means points are closer to their cluster centers.
```

```python
11. What is the elbow method in K-Means clustering?

Ans- The elbow method helps choose the right number of clusters in K-Means. You plot inertia (sum of distances) for different cluster counts and
     look for a point where the decrease slows down, forming an “elbow.” That point is the best number of clusters.
```

```python
12. Describe the concept of "density" in DBSCAN.

Ans- In DBSCAN, "density" means how many points are packed close together in a small area. If enough points are within a set distance (epsilon), that area is dense and can form a cluster. Low-density areas are considered noise or borders.
```

```python
13. Can hierarchical clustering be used on categorical data?

Ans- Yes, hierarchical clustering can be used on categorical data, but you need to use a distance measure that works for categories (like Hamming distance) instead of regular numeric distances.
```

```python
14. What does a negative Silhouette Score indicate?

Ans- A negative Silhouette Score means a point is closer to points in another cluster than to its own cluster, suggesting wrong or poor clustering.
```

```python
15. Explain the term "linkage criteria" in hierarchical clustering?

Ans- Linkage criteria in hierarchical clustering decide how to measure the distance between clusters when merging them. Examples are single linkage (closest points), complete linkage (farthest points), and average linkage (average distance).
```

```python
16. Why might K-Means clustering perform poorly on data with varying cluster sizes or densities?

Ans- K-Means assumes clusters are similar in size and density. If clusters vary, K-Means may group points incorrectly because it uses distance to the center and doesn't handle uneven shapes or densities well.
```

```python
17. What are the core parameters in DBSCAN, and how do they influence clustering?

Ans- The core parameters in DBSCAN are:
* Epsilon (ε) – the radius to search for neighbors. A small ε means tighter clusters.
* MinPts – the minimum number of points required to form a dense region (cluster).
```

```python
18. How does K-Means++ improve upon standard K-Means initialization?

Ans- K-Means++ improves K-Means by choosing initial cluster centers more smartly. It spreads them out by picking the first center randomly, then choosing the next centers far from the ones already chosen. This helps avoid poor clustering and speeds up convergence.
```

```python
19.  What is agglomerative clustering?

Ans- Agglomerative clustering is a type of hierarchical clustering that starts with each data point as its own cluster and then merges the closest clusters step by step until all points form one big cluster or a stopping condition is met.
```

```python
20. What makes Silhouette Score a better metric than just inertia for model evaluation?

Ans- Silhouette Score is better than just inertia because it considers both:
* How close a point is to its own cluster (cohesion)
* How far it is from other clusters (separation)
```

```python
#21. Generate synthetic data with 4 centers using make_blobs and apply K-Means clustering. Visualize using a scatter plot.

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

kmeans = KMeans(n_clusters=4, random_state=0)
y_kmeans = kmeans.fit_predict(X)

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='red', marker='X', label='Centers')
plt.title("K-Means Clustering with 4 Centers")
plt.legend()
plt.show()
```

```python
#29. Create synthetic data using make_blobs and evaluate silhouette scores for k = 2 to 5. Display as a bar chart.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

silhouette_scores = []

for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    score = silhouette_score(X, labels)
    silhouette_scores.append(score)

plt.bar(range(2, 6), silhouette_scores, color='skyblue')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for K-Means Clustering')
plt.show()
```

```python
#30. Load the Iris dataset and use hierarchical clustering to group data. Plot a dendrogram with average linkage.

import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data

Z = linkage(X, method='average')

plt.figure(figsize=(10, 5))
dendrogram(Z, truncate_mode='level', p=30)
plt.title('Hierarchical Clustering Dendrogram (Average Linkage)')
plt.xlabel('Sample Index or Cluster Size')
plt.ylabel('Distance')
plt.show()
```

```python
#31. Generate synthetic data with overlapping clusters using make_blobs, then apply K-Means and visualize with decision boundaries.

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from matplotlib.colors import ListedColormap

X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=2.5, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

h = 0.1  # mesh step size
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))

Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, cmap=cmap_light, alpha=0.4)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap=cmap_bold, edgecolor='k', s=40)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
            s=200, c='yellow', marker='X', label='Centroids')
plt.title("K-Means Clustering with Decision Boundaries (Overlapping Clusters)")
plt.legend()
plt.show()
```

```python
#44. Load the Iris dataset and plot the Silhouette Coefficient for each sample after KMeans clustering.

import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score

iris = load_iris()
X = iris.data

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

silhouette_vals = silhouette_samples(X, labels)
avg_score = silhouette_score(X, labels)

y_ticks = []
start = 0
plt.figure(figsize=(8, 6))

for i in range(3):
    cluster_silhouette_vals = silhouette_vals[labels == i]
    cluster_silhouette_vals.sort()
    y_lower = start
    y_upper = start + len(cluster_silhouette_vals)
    plt.barh(range(y_lower, y_upper), cluster_silhouette_vals, edgecolor='none', height=1)
    y_ticks.append((y_lower + y_upper) / 2)
    start = y_upper

plt.axvline(avg_score, color="red", linestyle="--", label=f'Avg Silhouette Score: {avg_score:.2f}')
plt.yticks(y_ticks, [f'Cluster {i}' for i in range(3)])
plt.xlabel('Silhouette Coefficient')
plt.ylabel('Cluster')
plt.title('Silhouette Coefficient per Sample for KMeans Clustering on Iris')
plt.legend()
plt.show()
```

```python
#45. Generate synthetic data using make_blobs and apply Agglomerative Clustering with 'average' linkage. Visualize clusters.

import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering

X, _ = make_blobs(n_samples=300, centers=4, cluster_std=1.2, random_state=42)

agg_clust = AgglomerativeClustering(n_clusters=4, linkage='average')
labels = agg_clust.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Set2', s=50)
plt.title('Agglomerative Clustering with Average Linkage')
plt.show()
```

```python
#46. Load the Wine dataset, apply KMeans, and visualize the cluster assignments in a seaborn pairplot (first 4 features).

import seaborn as sns
import pandas as pd
from sklearn.datasets import load_wine
from sklearn.cluster import KMeans

wine = load_wine()
X = wine.data[:, :4]  # first 4 features

kmeans = KMeans(n_clusters=3, random_state=42)
labels = kmeans.fit_predict(X)

df = pd.DataFrame(X, columns=wine.feature_names[:4])
df['Cluster'] = labels.astype(str)  # as categorical for color

sns.pairplot(df, hue='Cluster', palette='Set1')
import matplotlib.pyplot as plt
plt.suptitle('KMeans Clusters on Wine Dataset (First 4 Features)', y=1.02)
plt.show()
```

```python
#47. Generate noisy blobs using make_blobs and use DBSCAN to identify both clusters and noise points. Print the count.

from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np

X, _ = make_blobs(n_samples=500, centers=3, cluster_std=1.5, random_state=42)

dbscan = DBSCAN(eps=0.8, min_samples=5)
labels = dbscan.fit_predict(X)

n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = np.sum(labels == -1)

print(f'Number of clusters found: {n_clusters}')
print(f'Number of noise points: {n_noise}')
```

```python
#48. Load the Digits dataset, reduce dimensions using t-SNE, then apply Agglomerative Clustering and plot the clusters.

import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
from sklearn.cluster import AgglomerativeClustering

digits = load_digits()
X = digits.data

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)

agg_clust = AgglomerativeClustering(n_clusters=10)
labels = agg_clust.fit_predict(X_tsne)

plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=labels, cmap='tab10', s=50)
plt.title('Agglomerative Clustering on t-SNE Reduced Digits Data')
plt.xlabel('t-SNE Dim 1')
plt.ylabel('t-SNE Dim 2')
plt.show()
```
